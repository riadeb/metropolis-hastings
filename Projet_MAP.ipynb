# Algorithme de Metropolis-Hastings

**T1.** Vérifier que $P(x,y)$ est bien définie $\forall x,y \in E $. Montrer que

$$
P(x,y) = \left\{
    \begin{array}{ll}
        q(y)R(x,y) & \mbox{si} & x \neq y\\
        1 -  \sum_{z \neq x}q(z)R(x,z) & \mbox{sinon}
    \end{array}
\right.\\
$$

On sait que $P(X_0 = x) = \mu(x) > 0$, donc $P(x,y) = P(X_1 = y|X_0=x)$ est bien définie.


On a  
\begin{equation} 
(X_1 = y) \cap (X_0 = x) = (\left(   (X_1 = y) \cap (U_1 < R(X_0,Y_1)) \right) \cup \left(   (X_1 = y) \cap (U_1 \geq R(X_0,Y_1)) \right)) \cap (X_0 = x)
\end{equation}


Or, \begin{equation}  (X_1 = y) \cap (U_1 < R(X_0,Y_1) )  \cap (X_0 = x) = (Y_1 =y) \cap (U_1 < R(x_0,y_1) )  \cap (X_0 = x)\end{equation}

Et comme les variables aléatoires $Y_1,X_0 \,et \, U_1$ sont indépendantes, alors :

\begin{equation}
\mathbb{P}( (X_1 = y) \cap (U_1 < R(X_0,Y_1) )  \cap (X_0 = x)) = \mathbb{P}(Y_1 =y) \mathbb{P} (U_1 < R(x_0,y_1) )  \mathbb{P} (X_0 = x) = q(y)R(x,y)\mu(x)
\end{equation}

D'un autre côté, 

\begin{equation} 
(X_1 = y) \cap (U_1 \geq R(X_0,Y_1))  \cap (X_0 = x) = (X_0 = y) \cap (U_1 \geq R(X_0,Y_1))  \cap (X_0 = x)
\end{equation}

Donc si $x \neq y$ alors 

\begin{equation}
 \mathbb{P}((X_1 = y) \cap (U_1 \geq R(X_0,Y_1))  \cap (X_0 = x)) = 0
\end{equation} 

Ce qui montre que 

\begin{equation}
 \mathbb{P}((X_1 = y) \cap (X_0 = x)) = q(y)R(x,y)\mu(x)
\end{equation}

et donc que 

 \begin{equation}
P(x,y) =  \frac{\mathbb{P}((X_1 = y) \cap (X_0 = x))}{\mu(x)} =q(y)R(x,y)
\end{equation}

Et si $x=y$ alors : 

\begin{align}
 \mathbb{P}((X_1 = y) \cap (U_1 \geq R(X_0,Y_1))  \cap (X_0 = x)) &=  \mathbb{P}((X_0 = x) \cap (U_1 \geq R(X_0,Y_1))) \\
 &= \sum_{z \in E} \mathbb{P}((X_0 = x) \cap (U_1 \geq R(X_0,Y_1))\cap (Y_1 = z))\\
 &= \sum_{z \in E} \mathbb{P}((X_0 = x) \cap (U_1 \geq R(x,z))\cap (Y_1 = z)) \\
 &= \sum_{z \in E} \mu(x)q(z)(1-R(x,z)) \\
 &= \mu(x) - \sum_{z \in E} \mu(x)q(z)R(x,z)
 \end{align}
 
 Donc 
 
 \begin{align}
 \mathbb{P}((X_1 = y) \cap (X_0 = x)) &= q(y)R(x,y)\mu(x) + \mu(x) - \sum_{z \in E} \mu(x)q(z)R(x,z) \\
 &= \mu(x) - \sum_{z \neq x}q(z)R(x,z)\mu(x)
\end{align}

Et on obtient finalement

\begin{equation}
P(x,y) =  \frac{\mathbb{P}((X_1 = y) \cap (X_0 = x))}{\mu(x)} = 1 -  \sum_{z \neq x}q(z)R(x,z)
\end{equation}

**T2.** Montrer que, pour tout $x, y \in E$ on a \\

\begin{equation}
\pi(x)P(x, y) = \pi(y)P(y, x)
\end{equation}

Si $x \neq y :$
\begin{align}
\pi(x)P(x, y) &= \pi(x)q(y)R(x, y) \\
&=\pi(x)q(y)min( \frac{\pi(y)q(x)}{\pi(x)q(y)}, 1)\\
&= min( \pi(y)q(x),\pi(x)q(y)) \; car \; \pi(x)q(y) \geq 0\\
&= \pi(y)q(x)min( 1,\frac{\pi(x)q(y)}{\pi(y)q(x)})\\
&= \pi(y)q(x)R(y, x)\\
&= P(y, x) \pi(y)\\
\end{align}
Enfin si $x = y  $ c'est evident.  \\
**T3.** Montrer que pour tout $ y \in E$ on a \\
\begin{equation}
\mathbb{P}(X_1= y) = \underset{x \in E}{\sum}\mu(x) P(x, y)
\end{equation}
De façon génerale, montrer que 
\begin{equation}
\mathbb{P}(X_{n+1}= y) = \underset{x \in E}{\sum}\mathbb{P}(X_n = x)P(x, y) \\
\end{equation}
On calcule : \\
\begin{align}
\mathbb{P}(X_1= y) &= \underset{x \in E}{\sum}\mathbb{P}(X_0 = x)\mathbb{P}(X_1= y| X_0 = x) \\
&=\underset{x \in E}{\sum}\mu(x)P(x, y) \\
\end{align}
De façon génerale, on a :
\begin{align}
\mathbb{P}(X_{n+1}= y) &= \underset{x \in E}{\sum}\mathbb{P}(X_n = x)\mathbb{P}(X_{n+1} = y | X_n = x) \\
\end{align}
on montre que pour tout $n \in \mathbb{N} :$ 

\begin{equation}
\mathbb{P}(X_{n+1} = y | X_n = x) = \mathbb{P}(X_1 = y | X_0 = x) = P(x, y)
\end{equation}

on a si $x \neq y : $
\begin{align}
\mathbb{P}(X_{n+1} = y \cap X_n = x) &= \mathbb{P}(X_{n+1}= y\cap X_n = x \cap U_{n+1} \lt R(x,Y_{n+1})) +\mathbb{P}(X_{n+1} = y \cap X_n = x \cap U_{n+1} \geq R(x,Y_{n+1}))  \\
&= \mathbb{P}(Y_{n+1}= y \cap X_n = x \cap U_{n+1} \lt R(x,Y_{n+1}))+\mathbb{P}(X_{n} = y \cap X_n = x \cap U_{n+1} \geq R(x,Y_{n+1}) )  \\
&= \mathbb{P}(Y_{n+1}= y \cap X_n = x \cap U_{n+1} \lt R(x,y)) \:puisque\:x \neq y\: et \: Y_{n+1} \perp \!\!\! \perp X_n \\
&= \mathbb{P}(Y_{n+1}= y \cap X_n = x \cap U_{n+1} \lt R(x,y))\\
&= \mathbb{P}(Y_{n+1}= y)\mathbb{P} ( X_n = x )\mathbb{P}( U_{n+1} \lt R(x,y))\\
&= q(y)R(x,y)\mathbb{P} ( X_n = x )
\end{align}
donc : 
\begin{align}
\mathbb{P}(X_{n+1} = y | X_n = x) &=q(y)R(x,y)\\
&=P(x, y)\\
\end{align}

Sinon si $x = y : $
\begin{align}
\mathbb{P}(X_{n+1} = x \cap X_n = x) &= \mathbb{P}(X_{n+1}= x \cap X_n = x \cap U_{n+1} \lt R(x,Y_{n+1}))+\mathbb{P}(X_{n+1} = x \cap X_n = x \cap U_{n+1} \geq R(x,Y_{n+1})) \\
&= \mathbb{P}(Y_{n+1}= x \cap X_n = x \cap U_{n+1} \lt R(x,Y_{n+1}))+\mathbb{P}(X_{n} = x \cap X_n = x \cap U_{n+1} \geq R(x,Y_{n+1})) \\
&= \mathbb{P}(Y_{n+1}= x \cap X_n = x \cap U_{n+1} \lt R(x,x))+\mathbb{P}(X_{n} = x  \cap U_{n+1} \geq R(x,Y_{n+1})) \\
&=\mathbb{P}(Y_{n+1}= x)\mathbb{P} ( X_n = x )\mathbb{P}( U_{n+1} \lt R(x,x)) + \mathbb{P} ( X_n = x )\mathbb{P}(U_{n+1} \geq R(x,Y_{n + 1})) \\
\end{align}
et d'autre part : 
\begin{align}
\mathbb{P}(U_{n+1} \geq R(x,Y_{n+1})) & = \underset{z \in E}{\sum}\mathbb{P}(U_{n+1} \geq R(x,z))\mathbb{P}(Y_{n+1} = z)\\
&= \underset{z \in E}{\sum}(1- R(x,z))q(z)\\
&= 1 -\underset{z \in E}{\sum} R(x,z)q(z)\\
\end{align}
donc \
\begin{align}
\mathbb{P}(X_{n+1} = x | X_n = x) &=q(x)R(x,x) +1 -\underset{z \in E}{\sum} R(x,z)q(z) \\
&= 1 -\underset{z \neq x}{\sum} R(x,z)q(z)\\
&=P(x, x)\\
\end{align}
\\
On reprend :
\begin{align}
\mathbb{P}(X_{n+1}= y) &= \underset{x \in E}{\sum}\mathbb{P}(X_n = x)\mathbb{P}(X_{n+1} = y | X_n = x) \\
&= \underset{x \in E}{\sum}\mathbb{P}(X_n = x)P(x, y) \\
\end{align}
qui s'ecrira avec la notation de la loi de $X_n$ comme suit : 
\begin{align}
\pi_{n+1}( y) &= \underset{x \in E}{\sum}\pi_n( x)P(x, y)  \: pour \: tout \:y \\
\end{align}
**T4.** 
Montrer que si $\mu = \pi$ alors $\pi_n = \pi$ \
on procede par recurrence, \
pour $n = 0$ : $\pi_0 = \mu = \pi$ \
supposons que $\pi_n = \pi$ \
on a $ \forall y \in E $ :\
\begin{align}
\pi_{n+1}( y) &= \underset{x \in E}{\sum}\pi_n( x)P(x, y) \\
&=\underset{x \in E}{\sum}\pi( x)P(x, y)\\
&=\underset{x \in E}{\sum}\pi(y)P(y, x)\\
&=\pi(y)\underset{x \in E}{\sum}P(y, x)\\
&=\pi(y)
\end{align}
Donc $\pi_{n+1} = \pi$

**T5.** Montrer que 
$$ \forall x,y  \in  E \quad P(x,y) - K^{-1}\pi(y) \geq 0 $$

Soit $x,y \in E$. On a $K^{-1} = \underset{z \in E}{inf} \frac{q(x)}{\pi(x)} $ donc $K^{-1} \leq \frac{q(x)}{\pi(x)}$ , alors $K^{-1}\pi(y) \leq \frac{q(x)}{\pi(x)}\pi(y) =\frac{q(x)\pi(y)}{\pi(x)q(y)}q(y) $ et $K^{-1}\pi(y) \leq q(y) $.

On en déduit que $K^{-1}\pi(y) \leq min (\frac{q(x)\pi(y)}{\pi(x)q(y)}q(y), q(y)) = q(y)R(x,y)$

Donc si $x \neq y $ : $P(x,y) = q(y)R(x,y) \geq K^{-1}\pi(y)$

Et
\begin{align}
P(x,x) &= 1 -  \sum_{z \neq x}q(z)R(x,z) \\
&\geq 1 -  \sum_{z \neq x}q(z) = q(x) \\
&\geq K^{-1}\pi(x) 
\end{align}

Donc $$ \forall x,y  \in  E \quad P(x,y) - K^{-1}\pi(y) \geq 0 $$


**T6.** Montrer que $\forall n \geq 0, \quad d_{vt}(\pi_{n+1},\pi) = \frac{1}{2}\sum_y |\sum_x (\pi_n(x) - \pi(x))(P(x,y) - K^{-1}\pi(y)) |$

On a \begin{align}
d_{vt}(\pi_{n+1},\pi) &= \frac{1}{2}\sum_y|\pi_{n+1}(y) - \pi(y)| \\
\end{align}
Et 
 \begin{align}
\pi_{n+1}(y) - \pi(y) &= \sum_x \pi_n(x)P(x,y)  \quad - \pi(y)\\
&=  \sum_x \pi_n(x)P(x,y) \quad - \pi(y)\sum_xP(y,x) +  \sum_x K^{-1}\pi_n(x)\pi(y)  - \sum_x K^{-1}\pi_n(x)\pi(y) \\
&= \sum_x \pi_n(x)P(x,y) \quad - \sum_xP(x,y)\pi(x) +  \sum_x K^{-1}\pi(x)\pi(y)  - \sum_x K^{-1}\pi_n(x)\pi(y) \\
&= \sum_x \pi_n(x)(P(x,y) - K^{-1}\pi(y))+\pi(x)(K^{-1}\pi(y) - P(x,y)) \\
&=  \sum_x (\pi_n(x) - \pi(x))(P(x,y) - K^{-1}\pi(y)) \\
\end{align}

On conclut donc que $$ d_{vt}(\pi_{n+1},\pi) = \frac{1}{2}\sum_y|\pi_{n+1}(y) - \pi(y)| =   \frac{1}{2}\sum_y |\sum_x (\pi_n(x) - \pi(x))(P(x,y) - K^{-1}\pi(y)) |$$

**T7.** Montrer que $\forall n \geq 0, \quad d_{vt}(\pi_{n+1},\pi) \leq  d_{vt}(\pi_{n},\pi)(1-K^{-1}) $

D'aprés *T6*, on a 
\begin{align}
\quad d_{vt}(\pi_{n+1},\pi) &\leq \frac{1}{2}\sum_y \sum_x |\pi_n(x)-\pi(x)|P(x,y) - K^{-1}\pi(y)| \\
&=\frac{1}{2} \sum_y \sum_x |\pi_n(x)-\pi(x)|(P(x,y) - K^{-1}\pi(y)) \quad d'aprés \, T5\\
&=  \frac{1}{2}\sum_x |\pi_n(x)-\pi(x)|\sum_y (P(x,y) - K^{-1}\pi(y)) \\
&=  \frac{1}{2}\sum_x |\pi_n(x)-\pi(x)|(1-K^{-1}) \\
&= d_{vt}(\pi_{n},\pi)(1-K^{-1}) 
\end{align}

D'où le résultat.
